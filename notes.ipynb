## ðŸ” What is a Return Statement?

The `return` statement is used to **exit a function** and **send a value back** to the caller. It allows a function to produce a result that can be used elsewhere in the program.

### Syntax:
```python
def function_name():
    # some code
    return value
```

---

## ðŸ“Š Breakdown of Functions in the Code

### 1. Function `add()` - A Return Function

```python
def add():
    a = 2
    b = 3
    total = a + b
    return total
```

| Component | Description |
|-----------|-------------|
| `def add():` | Defines a function named `add` with no parameters |
| `a = 2` | Local variable assignment |
| `b = 3` | Local variable assignment |
| `total = a + b` | Calculates the sum (2 + 3 = 5) |
| `return total` | **Returns the value 5** to the caller |

**Return Type:** `int` (integer) - because `total` holds an integer value (5)

### 2. Function `print_result()` - A Void Function (Implicit Return)

```python
def print_result():
    result = add()
    print(f" The result is {result}")
```

| Component | Description |
|-----------|-------------|
| `def print_result():` | Defines a function with no parameters |
| `result = add()` | Calls `add()` and stores the returned value (5) in `result` |
| `print(...)` | Prints the result to console |
| No `return` statement | **Implicitly returns `None`** |

**Return Type:** `None` (void function)

## ðŸ“š Key Concepts of Return Functions

### 1. Purpose of `return`
- **Sends a value back** to the calling code
- **Terminates** the function execution immediately
- Allows functions to be used in **expressions** and **assignments**

### 2. Types of Functions Based on Return

| Type | Description | Example |
|------|-------------|--------|
| **Return Function** | Explicitly returns a value | `add()` returns `5` |
| **Void Function** | No return statement (returns `None` implicitly) | `print_result()` |

### 3. What Can Be Returned?
- **Single values:** integers, floats, strings, booleans
- **Multiple values:** using tuples
- **Collections:** lists, dictionaries, sets
- **Objects:** custom class instances
- **Other functions:** for functional programming

## ðŸ’¡ Examples of Return Functions

````python
# Example 1: Returning a single value
def square(n):
    return n * n

result = square(4)
print(f"Square: {result}")  # Output: Square: 16
````

````python
# Example 2: Returning multiple values (tuple)
def get_min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = get_min_max([5, 2, 8, 1, 9])
print(f"Min: {minimum}, Max: {maximum}")  # Output: Min: 1, Max: 9
````

````python
# Example 3: Early return (conditional return)
def is_positive(num):
    if num > 0:
        return True
    return False

print(is_positive(5))   # Output: True
print(is_positive(-3))  # Output: False
````

````python
# Example 4: Returning a list
def get_even_numbers(limit):
    return [x for x in range(limit) if x % 2 == 0]

evens = get_even_numbers(10)
print(f"Even numbers: {evens}")  # Output: Even numbers: [0, 2, 4, 6, 8]
````

## ðŸŽ¯ Type Hints (Modern Python)

In modern Python (3.5+), you can specify **return type hints** to make your code more readable and type-safe:

```python
def add() -> int:
    a = 2
    b = 3
    total = a + b
    return total

def print_result() -> None:
    result = add()
    print(f" The result is {result}")
```

| Annotation | Meaning |
|------------|--------|
| `-> int` | Function returns an integer |
| `-> None` | Function returns nothing (void) |
| `-> str` | Function returns a string |
| `-> list` | Function returns a list |
| `-> bool` | Function returns a boolean |

## âš ï¸ Common Mistakes with Return

### 1. Forgetting to return a value
```python
def add_wrong():
    total = 2 + 3
    # Missing return statement!

result = add_wrong()  # result will be None
```

### 2. Code after return (unreachable code)
```python
def example():
    return 5
    print("This will never execute!")  # Unreachable
```

### 3. Not using the returned value
```python
add()  # Value 5 is returned but discarded
# Better: result = add()
```
